# Visualizamos los datos

## Rutas e imports

```python
from pathlib import Path
import pandas as pd
from pyspark.sql.functions import col, to_date
from IPython.display import display

PROJECT_ROOT = Path("/home/math2data/nyc-mobility-lakehouse")

LANDING_DIR = PROJECT_ROOT / "data" / "landing"
BRONZE_DIR = PROJECT_ROOT / "data" / "bronze" / "yellow_taxi_trips"

print(PROJECT_ROOT)
print(LANDING_DIR.exists())
print(BRONZE_DIR.exists())
```

### Cargamos los parquets brutos

```python
landing_files = sorted(LANDING_DIR.glob("yellow_tripdata_2023-*.parquet"))
landing_files
```

### Visualizamos los parquets en bruto 

```python
for i, file in enumerate(landing_files, start=1):
    raw_df = pd.read_parquet(file)
    print(f"Visualizamos una muestra del Parquet {i}: {file.name}")
    display(raw_df.sample(5, random_state=42))
    print(f"Shape: {raw_df.shape}")
    
```

```python
raw_df.notna().sum()[raw_df.notna().sum() > 0]
```

```python
dtypes_por_archivo = {}

for file in landing_files:
    raw_df = pd.read_parquet(file)
    dtypes_por_archivo[file.name] = raw_df.dtypes
```

```python
# Convertimos a DataFrame: filas = columnas, columnas = archivos
comparativa_dtypes = pd.DataFrame(dtypes_por_archivo)

# Columnas donde no todos los archivos tienen el mismo dtype
dtypes_diferentes = comparativa_dtypes[
    comparativa_dtypes.nunique(axis=1, dropna=False) > 1
]

dtypes_diferentes

if dtypes_diferentes.empty:
    print("Todos los archivos tienen las mismas columnas con los mismos dtypes.")
else:
    print("Hay diferencias de dtypes o columnas entre archivos:")
    display(dtypes_diferentes)

```

```python
#En caso de que no haya los mismos tipos averiguamos donde se está produciendo el fallo
for columna, fila in dtypes_diferentes.iterrows():
    print(f"\nColumna con diferencias: {columna}")

    for archivo, dtype in fila.items():
        print(f"  {archivo}: {dtype}") 
```

## Visualizamos usando Spark

```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("View Landing and Bronze")
    .config("spark.driver.memory", "4g")
    .config("spark.jars.packages", "io.delta:delta-spark_2.12:3.1.0")
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension")
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog")
    .getOrCreate()
)

spark.sparkContext.setLogLevel("ERROR")
```

```python
bronze_df = spark.read.format("delta").load(str(BRONZE_DIR))
```

```python
bronze_df.limit(5).toPandas()
```

```python
bronze_df.filter(
    (col("year") == "2023") & (col("month") == "01")
).limit(5).toPandas()
```

```python
bronze_df.printSchema()
```

```python
bronze_df.groupBy("year", "month").count().orderBy("year", "month").toPandas()
```

```python
raw_columns = set(raw_df.columns)
bronze_columns = set(bronze_df.columns)

{
    "solo_en_raw": sorted(raw_columns - bronze_columns),
    
    "solo_en_bronze": sorted(bronze_columns - raw_columns),
    
    "comunes": sorted(raw_columns & bronze_columns)
}
```

```python
bronze_df.select(
    "source_file",
    "ingestion_timestamp",
    "service_type",
    "year",
    "month",
    "batch_id"
).limit(5).toPandas()
```

```python
from pyspark.sql import functions as F


raw_columns = set(raw_df.columns)
bronze_columns = set(bronze_df.columns)

comparativa_columnas = []

for colm in sorted(raw_columns | bronze_columns):
    comparativa_columnas.append({
        "columna": colm,
        "en_raw": colm in raw_columns,
        "en_bronze": colm in bronze_columns,
        "estado": (
            "común" if colm in raw_columns and colm in bronze_columns
            else "solo_en_raw" if colm in raw_columns
            else "solo_en_bronze"
        )
    })

comparativa_df = spark.createDataFrame(comparativa_columnas)

comparativa_df.orderBy("estado", "columna").toPandas()
```

```python
comparativa_df.filter(
    F.col("estado") != "común"
).orderBy("estado", "columna").toPandas()
```

```python
raw_lower = {c.lower(): c for c in raw_df.columns}
bronze_lower = {c.lower(): c for c in bronze_df.columns}

for col_lower in sorted(set(raw_lower) & set(bronze_lower)):
    if raw_lower[col_lower] != bronze_lower[col_lower]:
        print(
            f"Posible misma columna con distinto nombre: "
            f"raw='{raw_lower[col_lower]}', bronze='{bronze_lower[col_lower]}'"
        )
```

```python
dtypes_por_archivo_spark = {}

for file in landing_files:
    df_spark = spark.read.parquet(str(file))
    
    dtypes_por_archivo_spark[file.name] = {
        col_name: dtype
        for col_name, dtype in df_spark.dtypes
    }

comparativa_dtypes_spark = pd.DataFrame(dtypes_por_archivo_spark)

dtypes_diferentes_spark = comparativa_dtypes_spark[
    comparativa_dtypes_spark.nunique(axis=1, dropna=False) > 1
]

if dtypes_diferentes_spark.empty:
    print("Todos los archivos tienen las mismas columnas con los mismos dtypes.")
else:
    print("Hay diferencias de dtypes o columnas entre archivos:")
    display(dtypes_diferentes_spark)
```
